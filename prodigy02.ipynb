```python
#importing the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the dataset

data=pd.read_csv("Titanic-Dataset.csv")
data
```

```python
#shape of the dataset

data.shape
```

```python
#infomation about the dateset

data.info()
```

```python
#checking for the null values from the dataset

data.isnull()
```

```python
#finding the number of null values from the columns

data.isnull().sum()
```

```python
#dropping the unnecessary columns

data.drop(columns="Cabin",axis=1,inplace=True)
```

```python
#filling the missing values of the age column by mean of the age column

data['Age'].fillna(data['Age'].mean(),inplace=True)
```

```python
#filling the missing values of the embarked column by mean of the column

data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
#checking if any duplicate values are there are not

data.duplicated()
```

```python
#to know the number of duplicate values if any

data.duplicated().sum()
```

```python
#checking the count of people based on age

data['Age'].value_counts()
```

```python
#count of people survived

data['Survived'].value_counts()
```

```python
#representation of the people survived based on pie chart

plt.pie(data['Survived'].value_counts(),explode=[0,0.08],autopct="%1.2f%%",labels=['Not survived',' Survived'])
plt.title("Survival of people")
plt.show()
```

```python
#bar graph representing the survival counts

survival_counts = data['Survived'].value_counts()
plt.figure(figsize=(8, 6))
survival_counts.plot(kind='bar', color=['red', 'green'], alpha=0.75, edgecolor='black')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Counts')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.show()
```

```python
#COUNTPLOT
#visualization of the people survived from male and female


plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=data, palette=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survival Counts by Gender')
plt.legend(title='Survived', loc='upper right', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
#Male survival rate

data[data['Sex'] == 'male'].Survived.groupby(data.Survived).count().plot(kind='pie',
figsize=(5, 8),explode=[0,0.05],autopct='%1.1f%%',labels=["Not survived","Survived"])
plt.ylabel("")
plt.title("Male survival rate")
plt.show()
```

```python
#female survival rate

data[data['Sex'] == 'female'].Survived.groupby(data.Survived).count().plot(kind='pie',
figsize=(5, 8),explode=[0,0.05],autopct='%1.1f%%',labels=["Not survived","Survived"])
plt.ylabel("")
plt.title("Male survival rate")
plt.show()
```

```python
#visualization of people survived from different embarkment

sns.countplot(x='Embarked',data=data,palette='coolwarm',)
plt.xlabel("Embarked")
plt.ylabel("Number of people")
plt.show()
```

```python
#COUNTPLOT
#visualization of people survived from embarked column

plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', hue='Survived', data=data, palette=['green', 'yellow'])
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survival Counts by Embarked')
plt.legend(title='Survived', loc='upper right', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
#visualization of the people survived based on pclass

pclass_survival_counts = data.groupby('Pclass')['Survived'].value_counts().unstack()
colors = ['brown', 'red']
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
classes = pclass_survival_counts.index

for i, pclass in enumerate(classes):
    axes[i].pie(pclass_survival_counts.loc[pclass], labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=140, colors=colors)
    axes[i].set_title(f'Pclass {pclass} Survival Rate')

plt.suptitle('Survival Rate by Passenger Class')
plt.show()
```

```python
#BOXPLOT
#fare distribution by class

plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='pastel')
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
